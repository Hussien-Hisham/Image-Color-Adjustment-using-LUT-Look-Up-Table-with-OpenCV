```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

```python
# reads an input image
img = cv2.imread("..//content/OIP (3).jpeg", cv2.IMREAD_COLOR)
# converted it to 2D Grey img
image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# plotting histogram for the original 2D Grey img
cv2.imwrite("//content/OIP (3).jpeg",image)
```

```python
plt.hist(image)
plt.title("Image Histogram")
plt.xlabel("greyscale")
plt.ylabel("freq")
plt.figure()
plt.show()
```

```python
histr = cv2.calcHist([image],[0],None,[256],[0,256])
  
# show the plotting graph of an image
plt.plot(histr)
plt.show()
```

another way to draw histogram


```python
histr = cv2.calcHist([image],[0],None,[256],[0,256]
                     #img , chann , mask , histsize , ranges
plt.hist(image.ravel(),256,[0,256])
plt.title('Histogram for gray scale picture')
plt.show()
```

```python
plt.hist(image.ravel(),256,[0,256],cumulative=1)
plt.show()
```

```python


equ = cv2.equalizeHist(image)
res = np.hstack((image, equ))
  
# show image input vs output
cv2_imshow( res)
  
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.hist(equ)
plt.title("Image Histogram")
plt.xlabel("greyscale")
plt.ylabel("freq")
plt.figure()
plt.show()
```

```python
histr = cv2.calcHist([equ],[0],None,[256],[0,256])
plt.hist(equ.ravel(),256,[0,256])
plt.title('Histogram for gray scale picture')
plt.show()
```

```python
plt.hist(equ.ravel(),256,[0,256],cumulative=1)
plt.show()
```

```python
from google.colab.patches import cv2_imshow
img = cv2.imread('//content/OIP (3).jpeg')
original = img.copy()
xp = [0, 64, 128, 192, 255]
fp = [0, 16, 128, 240, 255]
x = np.arange(256)
table = np.interp(x, xp, fp).astype('uint8')
img = cv2.LUT(img, table)
cv2_imshow( original)
cv2_imshow( img)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```
